### importing required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
plt.style.use('dark_background')

#The warn() function defined in the ‘warning‘ module is used to show warning messages.
import warnings 
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("zomato.csv[1].zip")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df=df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
def handlerate(value):
    if(value=="NEW" or value=="-"):
        return np.nan
    else:
        value=str(value).split("/")
        value=value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'approx_cost(for two people)':'cost2plates','listed_in(type)':'type','listed_in(city)':'city'},inplace=True)
df.head()
```

```python
df['location'].unique()
```

```python
df['city'].unique()
```

```python
df=df.drop(['city'],axis=1)
```

```python
df.head()
```

```python
df['cost2plates'].unique()
```

```python
def handlecoma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
df['cost2plates']=df['cost2plates'].apply(handlecoma)
df['cost2plates'].head()
```

```python
df.head()
```

```python
rest_types=df['rest_type'].value_counts(ascending=False)
rest_types
```

```python
rest_types_lessthan1000=rest_types[rest_types<1000]
rest_types_lessthan1000
```

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
location1=df['location'].value_counts(ascending=False)
location1
```

```python
location_lessthan_300=location1[location1<300]
location_lessthan_300

def handle_locations(value):
    if(value in location_lessthan_300):
        return 'others'
    else:
        return value
df['location']=df['location'].apply(handle_locations)
df['location'].value_counts()
```

```python
df.head()
```

```python
df['cuisines'].value_counts()
```

```python
cuisines1=df['cuisines'].value_counts(ascending=False)


cuisines_lessthan_100=cuisines1[cuisines1<100]
```

```python


def handle_cuisines(value):
    if(value in cuisines_lessthan_100):
        return 'others'
    else:
        return value
df['cuisines']=df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

df.head()

```python
df['type'].value_counts()
```
